#### Matplotlib
- is the most popular plotting library for Python
- gives you control over ever aspect of a figure
- designed to have a similar feel to MatLab's graphical plotting

https://matplotlib.org/gallery/index.html

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y,'r-')
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
```

```python
plt.subplot(1,2,1) # numrows,numcols,plot#
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# Object-oriented method

fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8]) #[left,bottom,width,height]

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])

axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('LARGER PLOT')

axes2.plot(y,x)
axes2.set_title('SMALLER PLOT')
```

```python
fig,axes = plt.subplots(nrows=3,ncols=3,figsize=(8,4))

for i in axes:
    for axis in i:
        axis.plot(x,y)
        
plt.tight_layout()
```

#### Figure Size and DPI

```python
fig = plt.figure(figsize=(3,2)) # figsize=(width,height) - in inches

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig.savefig('my_picture.png',dpi=200)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2,label='Quadratic')
ax.plot(x,x**3,label='Cubic')

ax.legend(loc=10) # loc->center label
# or you can do ax.legend(loc=(0.1,0.1))
```

### Plot Appearance


```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,
        color='#FF8C00',
        linewidth=5,
        alpha=0.5,
        linestyle='--',
        drawstyle='steps',
        marker='o',
        markerfacecolor='blue',
        markersize=15) # color can accept hex codes, basic color name
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple',lw=2,ls='--')

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
