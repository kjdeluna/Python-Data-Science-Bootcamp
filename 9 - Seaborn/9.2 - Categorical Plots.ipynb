### Categorical Plot
Seeing the distributions of categorical columns (sex, day,time) in reference to either one of the numerical columns of another categorical column

```python
import seaborn as sns
%matplotlib inline

tips = sns.load_dataset('tips')
tips.head()
```

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
sns.countplot(x='sex',data=tips)
```

```python
# box - quartiles of dataset
# whiskers - show the rest of the distribution
# outside whiskers - outlier

sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',dodge=True) 
# jitter will insert random noise so the points are not stacked on top of each other
```

```python
# don't scale that well to very large numbers
sns.violinplot(x='day',y='total_bill',data=tips,color='gray')
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='bar')
```
